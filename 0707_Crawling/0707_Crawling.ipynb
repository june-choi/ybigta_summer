과제: 무신사 코디맵 페이지(많은 댓글 수 기준 정렬)에 나와있는 60개의 코디맵에 대하여 각각 ['codimap_category', 'codimap_title', 'codimap_date', 'views', 'comment_numbers'] 정보를 가져와서 csv 형태로 만든 다음, 해당 코드(.ipynb)와 csv 파일을 본인 github 과제 repository에 제출

```python
from bs4 import BeautifulSoup
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import numpy as np
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
print(selenium.__version__)
```

```python
driver = webdriver.Chrome('/Users/jeonghyeonchoi/Documents/ybigta_summer/chromedriver_mac_arm64/chromedriver')
```

```python
# 무신사 홈페이지
url = 'https://www.musinsa.com/app/codimap/lists?style_type=&tag_no=&brand=&display_cnt=60&list_kind=big&sort=comment_cnt&page=1'
driver.get(url)
driver.maximize_window() # 브라우저의 크기 전체화면으로 확대
```

```python
codimap_list = [] #코디맵 담을 리스트
#codimap_dic = {0:'기본'}
url = 'https://www.musinsa.com/app/codimap/lists?style_type=&tag_no=&brand=&display_cnt=60&list_kind=big&sort=comment_cnt&page=1' 
driver.get(url) 
driver.implicitly_wait(2)

columns = ['codimap_category', 'codimap_title', 'codimap_date', 'views', 'comment_numbers']    #행 이름 지정 
values = []                                             # 리스트 정보를 담을 빈 리스트 생성
soup = BeautifulSoup(driver.page_source, 'lxml')        # 현재 브라우저의 page의 html 정보를 lxml을 이용하여 가져오기 _ 이 아래부터는 뷰티풀 숩! 
data_rows = soup.find_all('li', attrs={'class':'style-list-item'}) # 해당 조건에 맞는 정보들을 list로 반환
```

```python
for i, row in enumerate(data_rows):                     # data_rows 리스트에 있는 값들과 인덱스를 차례대로 가져온다 _for 문으로 하는 이유는 두 번째 세 번째가 다 같은 형식으로 반복되기 때문! 
    print('{}번째 코디정보 크롤링'.format(i+1))
    blank = []

    codimap_category = row.find('span', attrs={'class':'style-list-information__text'})       # 코디분류 가져오기
    if codimap_category:                                                                    # 해당 조건에 맞는 정보가 하나라도 있을 경우:
        codimap_category = codimap_category.get_text().strip()                                          # html에서 텍스트만 가져온 뒤 공백 제거
        blank.append(codimap_category)
    else:                                                                       # 해당 조건에 맞는 정보가 하나도 없을 경우:
        blank.append('Something is wrong')
        print('{}번째 코디분류 가져올때 문제발생'.format(i+1))
        continue


    codimap_title = row.find('strong', attrs={'class':'style-list-information__title'})       # 코디이름 가져오기
    if codimap_title:                                                                    # 해당 조건에 맞는 정보가 하나라도 있을 경우:
        codimap_title = codimap_title.get_text().strip()                                          # html에서 텍스트만 가져온 뒤 공백 제거
        blank.append(codimap_title)
    else:                                                                       # 해당 조건에 맞는 정보가 하나도 없을 경우:
        blank.append('Something is wrong')
        print('{}번째 코디이름 가져올때 문제발생'.format(i+1))
        continue

    post_info = row.find('div', attrs={'class':'post-information'}) 
    post_info_txt = post_info.get_text().strip()
    post_info_lst = post_info_txt.split()

    codimap_date = post_info_lst[0]   # 코디날짜 가져오기
    if codimap_date:
        blank.append(codimap_date)
    else:
        blank.append('Something is wrong')
        print('{}번째 코디날짜 가져올때 문제발생'.format(i+1))
        continue

    views = post_info_lst[2]      # 조회수 가져오기
    if views:
        blank.append(views)
    else:
        blank.append('Something is wrong')
        print('{}번째 조회수 가져올때 문제발생'.format(i+1))
        continue

    comment_numbers = post_info_lst[4] # 댓글 가져오기
    if comment_numbers:
        blank.append(comment_numbers)
    else:
        blank.append('Something is wrong')
        print('{}번째 댓글 가져올때 문제발생'.format(i+1))
        continue

    values.append(blank)
    print('---------------------------------------------------')

time.sleep(1)

df = pd.DataFrame(values, columns = columns)
codimap_list.append(df)
```

```python
df_= pd.concat(codimap_list).reset_index(drop=True) # index 초기화 해줘야함
df_.head(10)
```
